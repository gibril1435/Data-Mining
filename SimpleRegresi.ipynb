# Read Data Set


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dskalori = pd.read_excel('kaloriberat.xlsx')
```

```python
dskalori
```

```python
dskalori.info()
```

```python
plt.scatter(dskalori['Kalori'], dskalori['Berat'])
plt.xlabel('Kalori')
plt.ylabel('Berat')
```

# Independent Dependent Variable

```python
x = dskalori.iloc[:,[0]].values
y = dskalori.iloc[:,[1]].values
```

```python
x
```

# Split Train Tets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_test
```

# Model Regeresi

```python
from sklearn.linear_model import LinearRegression
regresi = LinearRegression()
```

```python
regresi.fit(x_train,y_train)
```

# Visualisasi Data dan Garis Regresi

```python
# sebaran data train
plt.scatter(x_train,y_train)

# garis regresi
plt.plot(x_train,regresi.predict(x_train),c='red')
plt.xlabel('Kalori')
plt.ylabel('Berat')
```

# Testing Model

```python
y_prediksi = regresi.predict(x_test)
print(y_prediksi)
```

## Test Data Baru

```python
data_baru = [[650]]
prediksi_baru = regresi.predict(data_baru)
print(prediksi_baru)
```
